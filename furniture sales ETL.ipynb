```python
#Reading data and handling null values
import pandas as pd

# Use the correct path to the extracted CSV
df = pd.read_csv(r"C:\Users\Windows\Downloads\orders.csv")

# Preview the data
print(df.head(20))
```

```python
#types if ship mode - unique gives the unique values of shipment modes
df['Ship Mode'].unique()
```

```python
#replace these not avaible and unknown values as null /nan

df.replace(['Not Available', 'unknown'], pd.NA, inplace=True)
df['Ship Mode'].unique()
```

```python
#renaming the column names
df.columns
#change the column names to one pattern 
#1. use json and create a key value pair to replace each column manually
#2. instead use the python and convert the column names to str obj then to lower case then replace the spaces with underscore to acheive the default way
```

```python
#Changing the names of the columns to lower case
df.columns=df.columns.str.lower()
df.columns
```

```python
#replacing the spaces with underscore
df.columns = df.columns.str.replace(" ","_")
```

```python
df
```

```python
#list the discount amount
#in the data we have the discount percentage but not the exact value
df['discount']=df["list_price"]*df["discount_percent"]*.01
df['discount']
```

```python
df
```

```python
#find out the sale_price
df['sale_price']= df['list_price']-df['discount']
df
```

```python
#find out the total profit
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
#convert the date column which is in object to datetime
df['order_date']=pd.to_datetime(df['order_date'],format= "%Y-%m-%d")
```

```python
df.dtypes
```

```python
#drop columns which are not in use
df.drop(columns=['list_price','cost_price','discount_percent'],inplace = True,errors='ignore')
```
